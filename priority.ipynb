```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import google.generativeai as genai
import json
import fitz  # PyMuPDF for PDF extraction
```

```python
df = pd.read_csv("project_plans.csv")
```

```python
category_weightage = {
    "Healthcare": 10,
    "Infrastructure": 9,
    "Education": 8,
    "Water & Sanitation": 8,
    "Energy": 7,
    "Transport": 6,
    "Environment": 5,
    "Social Welfare": 4,
    "Tourism": 3,
    "IT & Digital Services": 3
}
```

```python
df["Category_Weight"] = df["Category"].map(category_weightage)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
df["Duration_Score"] = 1 - scaler.fit_transform(df[["Duration"]])
```

```python
df["Cost_Score"] = 1 - scaler.fit_transform(df[["Estimated_Cost (INR)"]])
```

```python
df["Priority_Score"] = df["Category_Weight"] * 0.5 + df["Duration_Score"] * 0.25 + df["Cost_Score"] * 0.25
```

```python
df = df.sort_values(by="Priority_Score", ascending=False)
```

```python
df.to_csv("prioritized_project_plans.csv", index=False)
```
